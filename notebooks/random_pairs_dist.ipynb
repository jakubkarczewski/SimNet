```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
```

```python
PL = True
```

```python
n = 50
```

```python
data = np.random.randint(100, size=(5000, 41))
data = data / 100
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
# compute pairwise cosine similarity
cos = cosine_similarity(df.values)

# conmstruct new DataFrame with cosine similarity
df_sim = pd.DataFrame(cos, columns=df.index.values, index=df.index.values)
df_sim.head()
```

```python
df.describe()
```

```python
# plot mean similarity distribution
sns.set_style('darkgrid')
data = df_sim.values.flatten()
sns.distplot(data)
plt.title('RozkÅ‚ad wzajemnego podobieÅ„stwa w zbiorze losowych wektorÃ³w' if PL else 
          'Distribution of pairwise similarity between random vectors')
plt.savefig('random_dist.png')

print(data.mean())
```

```python
movie_sims = dict()
for movie_name in df_sim.columns.values:
    movie_sims[movie_name] = dict()
    
    # choose most/least similar movies
    top_n = df_sim[movie_name].nlargest(n+1)
    low_n = df_sim[movie_name].nsmallest(n)
    
    # add them to our data along with similarity score
    movie_sims[movie_name]['pos'] = {k:v for k, v in zip(top_n.index.values[1:],
                                                         top_n.values[1:])}
    movie_sims[movie_name]['neg'] = {k:v for k, v in zip(low_n.index.values[1:],
                                                         low_n.values[1:])}

```

```python
sim_values = []
for movie_name in movie_sims.keys():
    for pair_type in movie_sims[movie_name].keys():
        values = list(movie_sims[movie_name][pair_type].values())
        sim_values.extend(values)    
        
sns.set_style('darkgrid')
sns.distplot(sim_values)
plt.title('RozkÅ‚ad wzajemnego podobieÅ„stwa kosinusowego \n dla par ze zbioru losowych wektorÃ³w' if PL else 
          'Distribution of similarity between pairs in random set')
plt.savefig('random_pairs_dist.png')
```
