```python
from time import time
import pickle

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
```

```python
with open('./sims.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
sim_values = []
for movie_name in data.keys():
    for pair_type in data[movie_name].keys():
        values = list(data[movie_name][pair_type].values())
        sim_values.extend(values)    
```

```python
len(sim_values)
```

```python
sns.set_style('darkgrid')
sns.distplot(sim_values)
plt.title('Distribution of similarity between pairs in training set')
plt.savefig('dist_pairs.png')
```
